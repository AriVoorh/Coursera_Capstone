```python
df=pd.read_csv('UNI_data_geo.csv')
```

```python
df['rank'].replace(regex=True,inplace=True,to_replace=r'=',value=r'')
```

```python
df = df.astype({"rank":'int'})
```

```python
# create map of New York using latitude and longitude values
map_uni = folium.Map(location=[46.2050242,6.1090692], zoom_start=4)

# add markers to map
for lat, lng, university in zip(df['Latitude'], df['Longitude'], df['University']):
    
    label = '{}'.format(university)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_uni)  
    
map_uni
```
